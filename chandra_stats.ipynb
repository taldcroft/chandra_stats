## Probability Distributions

#### Formally, a probability distribution assigns a probability to each outcome of an experiment.  
Examples:

- Flipping a coin once (discrete uniform distribution)
- Flipping a coin 10 times and counting heads (binomial distribition)
- Number of acquisition stars acquired in a star catalog (binomial distribution _[roughly]_)
- Clock time of radioactive decays modulo 1.0 sec (continuous uniform distribution [0, 1))
- Number of photons hitting a single ACIS pixel in one 3.3 sec readout (Poisson distribution)
- Total photons hitting ACIS in one 3.3 sec readout (Normal distribution)

For this context it is assumed that the "outcome of an experiment" is always a number.
The number could be an integer (discrete distribution) or a floating point value (continuous distribution).

A probability distribution can be sampled to create an ensemble of values.  If a large enough ensemble
is drawn from the distribution we can plot a *histogram* that represents the distribution.

**==> A probability distribution is like a normalized histogram <==**

### Plot the relative frequency of heads / tails for 5, 50, and 5000 coin flips

```python
def flip_coin(n_flips, p_heads=0.5):
    """
    Flip a coin ``n_flips`` times and return an array of
    the outcomes (True=heads, False=tails). The probability of
    getting heads is ``p_heads``.
    """
    # Generate ``n_flips`` random float values uniformly between 0 and 1
    uniform = np.random.uniform(0, 1, size=n_flips)
    heads = uniform < p_heads  # Set to True/False for each flip
    return heads
```

```python
p_heads = 0.5  # or 0.9 for example
for i, n_flip in enumerate([5, 50, 5000]):
    heads = flip_coin(n_flip, p_heads)
    subplot(1, 3, i + 1)
    hist(heads, bins=np.arange(-0.5, 2), normed=True, rwidth=0.5)
    ylim(0, 1)    
    title('{} flips'.format(n_flip))
    xticks([0, 1])
tight_layout()
```

### The [binomial distribution](http://en.wikipedia.org/wiki/Binomial_distribution)

In probability theory and statistics, the **binomial distribution** $b(n, p)$ is the discrete probability distribution of the number of successes in a sequence of $n$ independent yes/no experiments, each of which yields success with probability $p$.

[Wikipedia]

```python
n_trials = 5000
n_flip = 4
p_heads = 0.5  # or 0.9 for example
n_heads = []
for trial in range(n_trials):
    heads = flip_coin(n_flip, p_heads)
    n_heads.append(np.sum(heads))
print 'One sample of flips', heads
print 'Array of number of heads', n_heads[:20], '...'
```

```python
# Histogram of heads on top
# Scatter plot of n_heads for each trial on bottom
figure(figsize=(6, 8))
subplot(2, 1, 2)
plot(n_heads + np.random.uniform(-.1, .1, size=len(n_heads)), np.arange(n_trials), 
     '.', alpha=0.4, markersize=3)
ylabel('Trial number')
xlabel('Number of heads')
xlim(-1, n_flip + 1)
subplot(2, 1, 1)
# Plot the fraction of experiments resulting in the given number of heads
hist(n_heads, bins=np.arange(-0.5, n_flip + 1), normed=True, rwidth=0.25)

# Plot the corresponding binomial distribution
n_head_vals = [0, 1, 2, 3, 4]
frac_binomial = scipy.stats.binom(n_flip, 0.5).pmf(n_head_vals)
plot(n_head_vals, frac_binomial, 'or', markersize=10)
out = ylabel('Fraction of heads')
```

### Let's increase the number of tosses

```python
def count_heads(n_flips, n_trials=10000, p_heads=0.5):
    """
    Do ``n_trials`` of an experiment flipping a coin ``n_flips`` times
    with probability of heads ``p_heads``.  Return an array with the number 
    of heads for each of the trials (so the array is ``n_trials`` long).
    """
    n_heads = []
    for trial in range(n_trials):
        heads = flip_coin(n_flips, p_heads)
        n_heads.append(np.sum(heads))

    return np.array(n_heads)
```

```python
figure(figsize=(10, 4))
for n_flips in (40, 100, 200):
    n_heads = count_heads(n_flips)
    hist(n_heads, bins=np.arange(0, n_flips + 1))
    xlim(0, n_flips + 1)
xlabel('Number of heads')
out = xlim(0, 150)
```

### Use the [binomial distribution](http://en.wikipedia.org/wiki/Binomial_distribution) to investigate ACA acquisitions

**Background**: 
For each obsid we usually have a star catalog with 8 acquisition stars.  When we go to the target and the ACA acquires stars, between 0 and 8 of those stars will be acquired and identified by the OBC as good.

If the success of identification has a constant probability $p$, then the number of identified stars should follow the binomial distribution $b(8, p)$.

```python
acqs = np.load('acq_stats.npy')
print acqs[:2]
```

```python
# Compute the average success rate over all stars
print np.sum(acqs['obc_id'] == 'ID') / float(len(acqs))
```

```python
def count_acq_ids(acqs, obsid):
    """
    Given the acquisition dataset, count the number of successfully identified
    stars for the given ``obsid``.  Also return the total number of ACQ stars.
    """
    ok = acqs['obsid'] == obsid
    obs_acqs = acqs[ok]
    n_id = np.sum(obs_acqs['obc_id'] == 'ID')
    n_acq = len(obs_acqs)
    
    return n_id, n_acq
```

```python
# Make an array "n_ids" with the number of successful ACQs in each obsid.
# Only include obsids where there were 8 total ACQ in the catalog.

obsids = set(acqs['obsid'])  # get unique obsids
n_ids = []
for obsid in obsids:
    n_id, n_acq = count_acq_ids(acqs, obsid)
    if n_acq == 8:
        n_ids.append(n_id)
n_ids = np.array(n_ids)
```

#### Use [scipy.stats.binom](http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html) to compute the expected binomial distribution.

```python
# Plot a histogram of normalized success fractions for 0 to 8.
figure(figsize=(12,8))
hist(n_ids, bins=np.arange(-0.5, 9), normed=True, log=True, rwidth=0.5)

# Compute the corresponding binomial distribution for candidate values of "p"
for p_id in (0.86, 0.88, 0.90, 0.92, 0.94, 0.96):
    n = np.arange(9)  # 0, 1, ... 8
    p_n = scipy.stats.binom.pmf(n, 8, p_id)  # Compute probability of n successful acqs
    plot(n, p_n, label=str(p_id), linewidth=2)
    plot(n, p_n, '.', markersize=15)
ylim(.5e-5, 1)
xticks([0, 1, 2, 3, 4, 5, 6, 7, 8])
xlim(-0.5, 9)
legend(loc='best')
out = title('Successful star acqs in each catalog')
```

#### Conclusion: star acquisition is not well described by a simple binomial process with a single probability

### Play around with plotting and averaging distributions

```python
def average_samples(distr_func, n_samp, n_avg):
    """
    Return an ensemble of ``n_samp`` values taken by averaging 
    ``n_avg`` samples of the random distribution function ``distr_func``.
    """
    vals = distr_func(size=n_avg * n_samp)
    vals = vals.reshape(n_avg, n_samp)
    return np.mean(vals, axis=0)  # Array of length n_samp
```

```python
# Uniform distribution
n_samp = 10000
uniform_avg = average_samples(np.random.uniform, n_samp, 1) # try 2, 4, 10, 50, 100
out = hist(uniform_avg, bins=50, normed=True)
```

```python
# Exponential distribution
exp_avg = average_samples(np.random.exponential, n_samp, 1) # try 2, 4, 10, 50, 100
out = hist(exp_avg, bins=50, normed=True)
```

```python
# Normal (Gaussian) distribution
exp_avg = average_samples(np.random.normal, n_samp, 1) # try 2, 4, 10, 50, 100
out = hist(exp_avg, bins=50, normed=True)
```

### Interesting!  ANY distribution seems to converge to a normal (Gaussian) distribution after averaging enough.

This is the upshot of the famous [Central Limit Theorem](http://en.wikipedia.org/wiki/Central_limit_theorem).  

It is profoundly important because the Centra Limit Theorem is the whole reason that many measured quantities can be characterized using the normal (Gaussian) distribution.  

To be concrete, you compute the mean $\mu$ and standard deviation $\sigma$ of a sample of measurements.
Then you claim that 95% of measurements will be within $2-\sigma\ $ of the mean and 99.7% of measurements will be within $3-\sigma\ $ of the mean.
**This reasoning rests on the Central Limit Theorem.**

---

### If you have good evidence that a distribution is Normal (Gaussian) then the analytical expression can be used.

**Example**: compute the standard deviation of some values and compute the probability of seeing a value outside a certain limit.

```python
vals = np.random.normal(size=10000)  # Generate normally-distributed values
mean = np.mean(vals)  # Compute mean
std = np.std(vals)  # Compute standard deviation
print mean, std
```

```python
from scipy.stats import norm  # Everything you need from a normal distribution
x = np.linspace(-10, 10, 100)
plot(x, norm.cdf(x, loc=mean, scale=std))  # Cumulative distribution function -inf to x
hi = norm.cdf([1, 2, 3], loc=mean, scale=std)
lo = norm.cdf([-1, -2, -3], loc=mean, scale=std)
print hi - lo
```

```python
# Percentile point function (inverse of CDF)
print norm.ppf([0.01, 0.05, 0.50, 0.95, 0.99], loc=0.0, scale=1.0)
```

### The myth of $6-\sigma$

After averaging, the tails of a distribution converge to a Gaussian the slowest.  That means even if the central part of your sample seems pretty Gaussian, there is a very good chance the outer wings are **not** normally distributed.  Extrapolating beyond what you have measured in the data is asking for trouble.

$6-\sigma\ $ is about 1 / 500,000,000.  Unless you have around a billion independent samples, don't go there.

### Linear data fitting

This refers to fitting the data with a model that is linear in the model parameters.  This doesn't mean just fitting a line.  

Examples of linear models with free parameters `a`, `b`, `c`:

    y = a + b * x
    y = a + b * x + c * x**2
    y = a + b * x + c * sin(x / 200)

Examples of non-linear models:

    y = a * sin(c * x)
    y = a + b * exp(c * x)

The difference is that the linear models can be expressed as a linear matrix equation.

### Bootstrap resampling for parameter error estimation and extrapolation

#### Our view of the world on New Year's day 2010:

The IPS tank was well below its temperature or pressure limits, but what if we had asked:
_How hot can it go before the pressure exceeds 300 PSI?_

```python
x = prop['pftank2t'].vals
y = prop['pftankip'].vals
ok = prop.times < DateTime('2010:001').secs
xok = x[ok]
yok = y[ok]
plot(xok, yok, '.')
xlabel('Tank temperature (F)')
ylim(280, 302)
plot([70, 90], [300, 300], '--r')
out = ylabel('Tank pressure (PSI)')
```

```python
# This is the crux of bootstrap resampling: make a new dataset by
# picking randomly, WITH REPLACEMENT, from the set of original data points.

def resample(x, y):
    idxs = np.random.random_integers(0, len(x) - 1, size=len(x))
    return x[idxs], y[idxs]
```

```python
# Do 2000 trials of resampling to make a new dataset, fitting the
# line each time.  Store the slopes and intercepts.

ms = []  # slopes
bs = []  # intercepts
x_off = np.mean(xok) # Reparameterize the problem: good numerically and for analysis
n_sample = 2000
for i in range(n_sample):
    xr, yr = resample(xok, yok)
    m, b = polyfit(xr - x_off, yr, 1)
    ms.append(m)
    bs.append(b)
ms = np.array(ms)
bs = np.array(bs)
```

```python
figure(figsize=(12, 8))
plot(xok, yok, '.')
x0 = np.array([70, 110])
for m, b in zip(ms[:100], bs[:100]):
    y0 = m * (x0 - x_off) + b
    plot(x0, y0, alpha=0.2, color='r')
plot([70, 110], [300, 300], '--r')
out = xlim(70,110)
```

```python
out = plot(ms, bs - 280, '.')
xlabel('Slope')
out = ylabel('Intercept - 280')
```

```python
figure(figsize=(12, 5))
subplot(1, 2, 1)
hist(ms)
xlabel('Distribution of slopes')
subplot(1, 2, 2)
hist(bs - 283.7)
out = xlabel('Distribution of intercepts - 283.7')
tight_layout()
```

### Now to answer the original question: what temperature is will keep pressure < 300 PSI?

```python
# Run 10000 trials:
#
# - Randomly select one of the linear fits
#      This accounts for uncertainty in the slope/intercept
# - Randomly select one of the residual values (data - fit)
#      This accounts for the random noise in each point.
# - Randomly select a temperature (x) between 100 and 110 F

n_samp = 100000
line_idxs = np.random.randint(len(ms), size=n_samp) - 1
resid_idxs = np.random.randint(len(resid), size=n_samp) - 1
x_rand = np.random.uniform(100.0, 110.0, size=n_samp)
m_rand = ms[line_idxs]
b_rand = bs[line_idxs]
resid_rand = resid[resid_idxs]

# Now compute the expected pressure (y) for each line and random noise
y_rand = m_rand * (x_rand - x_off) + b_rand + resid_rand

# Plot the sampled values as a scatter plot
figure(figsize=(14, 6))
subplot(1, 2, 1)
plot(x_rand, y_rand, '.', alpha=0.5, markersize=0.5)
plot([100, 110], [300, 300], '--r', lw=2)

# Plot the line from the best fit using all available (recent) data
x0 = np.array([100, 110])
y0 = np.polyval(coeffs_all, x0)
plot(x0, y0, 'r', linewidth=3, label='Best fit with recent data')
legend(loc='upper left')

# Plot the values as a histogram for points within 0.25 F of 300 F.
subplot(1, 2, 2)
ok = np.abs(y_rand - 300) < 0.25
temp_at_300 = x_rand[ok]
hist(temp_at_300, bins=30, normed=True)
xlim(102, 107)
temp_percentiles = np.percentile(temp_at_300, [1, 5, 50, 95, 99])
y0, y1 = ylim()
for t in temp_percentiles:
    out = vlines(temp_percentiles, y0, y1, color='r')
```

### Sigma clipping

```python
def sigma_clip(y, y_fit, mask=None, n_sigma=3.0):
    """
    For the given ``y``, find points that are more than
    ``sigma`` standard-deviations from ``y_fit``.  Return a
    boolean ``mask`` containing good points.
    """
    if mask is None:
        mask = np.ones(len(x), dtype=np.bool)
    resid = y - y_fit
    std = np.std(resid[mask])
    ok = np.abs(resid / std) < n_sigma
    mask = mask & ok
    return mask
```

```python
# Make 3 passes of computing std deviation, masking bad points, and re-fitting

mask = None
for iter in range(3):
    mask = sigma_clip(y, y_fit, mask)
    coeffs = np.polyfit(x[mask], y[mask], 3)
    y_fit = np.polyval(coeffs, x)
plot_cxctime(elbv.times, y, alpha=0.4)
plot_cxctime(elbv.times, y_fit, 'r')
plot_cxctime(elbv.times[mask], y[mask], 'b')
ylabel('Load bus voltage')
grid()
```

### [Theil-Sen estimator](http://en.wikipedia.org/wiki/Theil%E2%80%93Sen_estimator) for line fits

Simple but clever way of rejecting data outliers in a line fit: take the median slope of all pairs of points (or an ensemble of randomly-selected pairs).

```python
# Fetch IPS tank temp and pressure since 2011
prop11 = fetch.Msidset(['pftank2t', 'pftankip'], '2011:001', stat='daily')
prop11.interpolate(86400)
x = prop11['pftank2t'].vals
y = prop11['pftankip'].vals
```

```python
# Inject some spurious data
x = np.concatenate((x, np.random.uniform(80, 95, size=50)))
y = np.concatenate((y, np.random.uniform(282, 290, size=50)))
coeffs = polyfit(x, y, 1)
y_fit = polyval(coeffs, x)
plot(x, y, '.')
plot(x, y_fit)
grid()
```

```python
def theil_sen(x, y):
    #
    # Input:
    #   x,y as linear numpy arrays
    # Output:
    #   b_0, b_1: estimates of intercept and slope
    #   res: residuals using the above linear model
    #
    # This method is discussed in the chapter "correlation and regression"
    # of the book "Basic Statistics: Understanding Conventional Methods" by
    # Rand R. Wilcox.
    #
    # Written by:
    # -- 
    # John L. Weatherwax                2009-04-21
    # 
    # email: wax@alum.mit.edu
    # 
    # Please send comments and especially bug reports to the
    # above email address.
    # 
    #-----

    n   = len(x)
    ord = numpy.argsort(x)
    xs  = x[ord]
    ys  = y[ord]
    vec1 = numpy.zeros( (n,n) )
    for ii in range(n):
        for jj in range(n):
            vec1[ii,jj] = ys[ii]-ys[jj]
    vec2 = numpy.zeros( (n,n) )
    for ii in range(n):
        for jj in range(n):
            vec2[ii,jj] = xs[ii]-xs[jj]
    v1    = vec1[vec2>0]    
    v2    = vec2[vec2>0]     
    slope = numpy.median( v1/v2 )
    coef  = numpy.zeros( (2,1) ) 
    b_0   = numpy.median(y)-slope*numpy.median(x)
    b_1   = slope
    res   = y-b_1*x-b_0 # residuals 

    return b_0, b_1, res
```

```python
b, m, resid = theil_sen(x, y)
y_fit = m * x + b
plot(x, y, '.')
plot(x, y_fit, 'r')
grid()

# NOT bad!  This can also be used as the starting point for sigma-clipping
```

## Non-linear model fitting

When the model is non-linear in the fitted parameters then the fitting problem is more difficult.

Reference: [Numerical Recipes](http://www.amazon.com/Numerical-Recipes-3rd-Scientific-Computing/dp/0521880688)

There are many ways to tackle this problem, but for our purpose it boils down to three things:

- Define a **model** which is a function of one or more **parameters**
- Define a **fit statistic** which estimates how well the model matches the data (smaller is better)
- Find the set of parameters that minimize the fit statistic

The [Sherpa modeling and fitting package](http://cxc.harvard.edu/sherpa/index.html) is a powerful tool for non-linear fitting and parameter estimation problems.  The learning curve can be a bit steep, but with local help (TLA) the basic methods can be mastered quickly.

```python
import numpy as np
import sherpa.ui as ui

def func(pars, x):
    ampl = pars[0]
    tau = pars[1]
    offset = pars[2]
    return ampl * np.exp(-x / tau) + offset
```

```python
x = np.linspace(0, 3, 50)
y = func([2.5, 0.75, 0.5], x)
sigma = 0.3
yn = y + sigma * np.random.normal(size=len(x))
```

```python
plot(x, y)
plot(x, yn, '.')
```

```python
# Define a dataset, providing the x, y and y uncertainty values.
yerr = sigma * np.ones_like(yn)
ui.load_arrays(1, x, yn, yerr)  # dataset 1

# Make a custom user model in sherpa
#  "func" is the name of the function that computes the model.
#  "fn" is the name of the sherpa model corresponding to "func".
#  These must be distinct.
ui.load_user_model(func, 'fn')
ui.add_user_pars('fn', ['ampl', 'tau', 'offset'])

# In the real world you typically want to define starting guess values
# and put in constraints that are reasonable to guide the fitting algorithm.
fn.tau = 1
fn.ampl = 1
fn.offset = 1
fn.tau.min = 0.05

# Set the model for dataset 1 to our "fn" model
ui.set_model(1, 'fn')
ui.plot_data(1)  # sherpa built-in method for plotting
```

```python
# Sherpa has many minimization algorithms.  Nelder-Mead (Simplex) is a good
# comprimise between accuracy and speed.
ui.set_method('neldermead')

# Now do the fit
ui.fit(1)
# The "Reduced statistic" should be around 1.0 (from 0.75 to 1.5 is probably OK)
```

```python
ui.plot_fit_resid(1)
```

```python
fit_results = ui.get_fit_results()
print fit_results
```

```python
parvals = fit_results.parvals
yfit = func(parvals, x)
plot(x, y)
plot(x, yn, '.')
out = plot(x, yfit)
```

```python
# The covariance matrix is an output of the minimization process that
# describes the uncertainty in the fitted parameters.  The diagonal elements
# are the variance (std dev squared) of each parameter while the off-diagonal
# elements describe the parameter correlations.

# For simple well-conditioned problems this is a good estimator for
# parameter uncertainty.

ui.covar()  # Get the computed covariance
covar = ui.get_covar_results().extra_output
print covar
```

```python
# The covariance matrix defines a multi-dimensional ellipsoid which
# can be sampled using the multivariate_normal() function.  

# This has roughly the same effect as bootstrap-resampling and refitting,
# but is *much* faster.  Non-linear fitting is generally slow compared
# to similar linear fits.

pars_list = np.random.multivariate_normal(parvals, covar, size=1000)
```

### Real world example: Fit the load bus voltage with a nonlinear function

```python
# Again, let's see how things looked on New Year's Day 2010

elbv = fetch.Msid('elbv', '2000:001', '2010:001', stat='daily')
plot_cxctime(elbv.times, elbv.vals)
out = ylabel('Load bus voltage')
```

```python
def exp_sin(pars, day):
    """
    Sherpa model function: decaying exponential with additive sinusoidal component
    """
    exp_ampl = pars[0]
    tau = pars[1]
    offset = pars[2]
    sin_ampl = pars[3]
    sin_offset = pars[4]
    phase = (day - sin_offset) * 2 * np.pi / 365.25
    return exp_ampl * np.exp(-day / tau) + offset + sin_ampl * np.sin(phase)
```

```python
# Set up the Sherpa dataset.  Reparameterize time value to days for convenience.
sigma = 0.1  # Educated guess on sigma
x = (elbv.times - elbv.times[0]) / 86400.0
y = elbv.vals
yerr = sigma * np.ones_like(y)
ui.load_arrays(1, x, y, yerr)

# Set up the Sherpa model
ui.load_user_model(exp_sin, 'es')
ui.add_user_pars('es', ['exp_ampl', 'tau', 'offset', 'sin_ampl', 'sin_offset'])
ui.set_source(1, 'es')
ui.plot_data(1)

# Initial guesses
es.exp_ampl = 1.0
es.tau = 1000.0
es.offset = 29.5
es.sin_ampl = 0.5
es.sin_offset = 90.0
```

```python
ui.fit()
```

```python
parvals = [par.val for par in es.pars]
y_fit = exp_sin(parvals, x)
figure(figsize=(12, 6))
subplot(1, 2, 1)
plot(x, y)
plot(x, y_fit, 'r', linewidth=2)
subplot(1, 2, 2)
plot(x, y_fit - y, '.')
```
