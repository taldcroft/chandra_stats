### Linear data fitting

This refers to fitting the data with a model that is linear in the model parameters.  This doesn't mean just fitting a line.  

Examples of linear models:

    y = a + b * x
    y = a + b * x + c * x**2
    y = a + b * x + c * sin(x / 200)

Examples of non-linear models:

    y = a * sin(c * x / 200)
    y = a + b * exp(c * x)

### Data outliers

Frequently the dataset to be fit has outliers, i.e. non-gaussian deviations that can distort the fit.

```python
elbv = fetch.Msid('elbv', '2009:100', '2009:250', stat='daily')
plot_cxctime(elbv.times, elbv.mins)
ylabel('Load bus voltage')
grid()
```

```python
x = (elbv.times - elbv.times[0]) / 86400  # days since start of interval
y = elbv.mins
coeffs = polyfit(x, y, 3)
plot_cxctime(elbv.times, y)
y_fit = polyval(coeffs, x)
plot_cxctime(elbv.times, y_fit, 'r')
ylabel('Load bus voltage')
grid()
```

#### Sigma clipping

```python
def sigma_clip(y, y_fit, mask=None, n_sigma=3.0):
    if mask is None:
        mask = np.ones(len(x), dtype=np.bool)
    resid = y - y_fit
    std = np.std(resid[mask])
    ok = np.abs(resid / std) < n_sigma
    mask = mask & ok
    return mask
```

```python
mask = None
for iter in range(3):
    mask = sigma_clip(y, y_fit, mask)
    coeffs = polyfit(x[mask], y[mask], 3)
    y_fit = polyval(coeffs, x)
plot_cxctime(elbv.times, y, alpha=0.4)
plot_cxctime(elbv.times, y_fit, 'r')
plot_cxctime(elbv.times[mask], y[mask], 'b')
ylabel('Load bus voltage')
grid()
```

### [Theil-Sen estimator](http://en.wikipedia.org/wiki/Theil%E2%80%93Sen_estimator) for line fits

Simple but clever way of rejecting data outliers in a line fit: take the median slope of all pairs of points (or an ensemble of randomly-selected pairs).

```python
prop11 = fetch.Msidset(['pftank2t', 'pftankip'], '2011:001', stat='daily')
prop11.interpolate(86400)
x = prop11['pftank2t'].vals
y = prop11['pftankip'].vals
```

```python
# Inject some spurious data
x = np.concatenate((x, np.random.uniform(80, 95, size=50)))
y = np.concatenate((y, np.random.uniform(282, 290, size=50)))
coeffs = polyfit(x, y, 1)
y_fit = polyval(coeffs, x)
plot(x, y, '.')
plot(x, y_fit)
grid()
```

```python
def theil_sen(x, y):
    #
    # Input:
    #   x,y as linear numpy arrays
    # Output:
    #   b_0, b_1: estimates of intercept and slope
    #   res: residuals using the above linear model
    #
    # This method is discussed in the chapter "correlation and regression"
    # of the book "Basic Statistics: Understanding Conventional Methods" by
    # Rand R. Wilcox.
    #
    # Written by:
    # -- 
    # John L. Weatherwax                2009-04-21
    # 
    # email: wax@alum.mit.edu
    # 
    # Please send comments and especially bug reports to the
    # above email address.
    # 
    #-----

    n   = len(x)
    ord = numpy.argsort(x)
    xs  = x[ord]
    ys  = y[ord]
    vec1 = numpy.zeros( (n,n) )
    for ii in range(n):
        for jj in range(n):
            vec1[ii,jj] = ys[ii]-ys[jj]
    vec2 = numpy.zeros( (n,n) )
    for ii in range(n):
        for jj in range(n):
            vec2[ii,jj] = xs[ii]-xs[jj]
    v1    = vec1[vec2>0]    
    v2    = vec2[vec2>0]     
    slope = numpy.median( v1/v2 )
    coef  = numpy.zeros( (2,1) ) 
    b_0   = numpy.median(y)-slope*numpy.median(x)
    b_1   = slope
    res   = y-b_1*x-b_0 # residuals 

    return b_0, b_1, res
```

```python
b, m, resid = theil_sen(x, y)
y_fit = m * x + b
plot(x, y, '.')
plot(x, y_fit, 'r')
grid()
```

## Non-linear fitting

```python
import numpy as np
import sherpa.ui as ui

def func(pars, x):
    ampl, inv_tau, offset = pars[0], pars[1], pars[2]
    return ampl * np.exp(-x * inv_tau) + offset
```

```python
x = np.linspace(0, 3, 50)
y = func([2.5, 1.3, 0.5], x)
sigma = 0.3
yn = y + sigma * np.random.normal(size=len(x))
```

```python
plot(x, y)
plot(x, yn, '.')
grid()
```

```python
yerr = sigma * np.ones_like(yn)
ui.load_arrays(1, x, yn, yerr)
ui.load_user_model(func, 'fn')
ui.add_user_pars('fn', ['ampl', 'inv_tau', 'offset'])
ui.set_source(1, 'fn')
ui.plot_data(1)
grid()
```

```python
ui.set_method('neldermead')
ui.fit(1)
# The "Reduced statistic" should be around 1.0 (from 0.75 to 1.5 is probably OK)
```

```python
elbv = fetch.Msid('elbv', '2000:001', '2010:001', stat='daily')
plot_cxctime(elbv.times, elbv.vals)
ylabel('Load bus voltage')
grid()
```

```python
def exp_sin(pars, day):
    exp_ampl = pars[0]
    tau = pars[1]
    offset = pars[2]
    sin_ampl = pars[3]
    sin_offset = pars[4]
    phase = (day - sin_offset) * 2 * np.pi / 365.25
    return exp_ampl * np.exp(-day / tau) + offset + sin_ampl * np.sin(phase)
```

```python
sigma = 0.1
x = (elbv.times - elbv.times[0]) / 86400.0
y = elbv.vals
yerr = sigma * np.ones_like(y)
ui.load_arrays(1, x, y, yerr)
ui.load_user_model(exp_sin, 'es')
ui.add_user_pars('es', ['exp_ampl', 'tau', 'offset', 'sin_ampl', 'sin_offset'])
ui.set_source(1, 'es')
ui.plot_data(1)
es.exp_ampl = 1.0
es.tau = 1000.0
es.offset = 29.5
es.sin_ampl = 0.5
es.sin_offset = 90.0
```

```python
ui.fit()
```

```python
parvals = [par.val for par in es.pars]
y_fit = exp_sin(parvals, x)
figure(figsize=(12, 6))
subplot(1, 2, 1)
plot(x, y)
plot(x, y_fit, 'r', linewidth=2)
subplot(1, 2, 2)
plot(x, y_fit - y, '.')
```
