```python
def count_heads(n_flips):
    n_heads = []
    for trial in range(10000):
        heads = np.random.random_integers(0, 1, size=n_flips)
        n_heads.append(np.sum(heads))

    hist(n_heads, bins=np.arange(0, n_flips+1), align='left')
    xlim(0, n_flips)
    grid()
    return n_heads
```

```python
n_heads = count_heads(100)
```

```python
import numpy as np
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
x = np.linspace(0, 2, 50)
y = func(x, 2.5, 1.3, 0.5)
yn = y + 0.4 * np.random.normal(size=len(x))
```

```python
plot(x, y)
plot(x, yn, '.')
```

```python
popt, pcov = curve_fit(func, x, yn)
print popt
print pcov
```

```python
yfit = func(x, *popt)
```

```python
plot(x, y)
plot(x, yn, '.')
plot(x, yfit)
```

```python
pars = np.random.multivariate_normal(popt, pcov, size=1000)
```

```python
pars.shape
```

```python
plot(pars[:, 1], pars[:, 2], '.')
```

```python
plot(x, yn, '.')
xpred = np.arange(0, 3, 0.1)
for par in pars[:100]:
    plot(xpred, func(xpred, par[0], par[1], par[2]), alpha=0.3, color='k')
```
